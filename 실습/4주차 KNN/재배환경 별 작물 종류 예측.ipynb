```python
import random
import os
seed = 42
random.seed(seed)
np.random.seed(seed)
os.environ["PYTHONHASHSEED"] = str(seed)
#데이터를 읽어오기
answer = pd.read_csv('/kaggle/input/2022-ml-w4p1/sample.csv')
train = pd.read_csv('/kaggle/input/2022-ml-w4p1/train.csv')
test = pd.read_csv('/kaggle/input/2022-ml-w4p1/test.csv')
```

```python
train
```

```python
# train과 test의 데이터 형태
# Dtype object는 문자열 의미
print(train.shape)
print(test.shape)
train.info()
test.info()
```

```python
# drop를 통해 인덱스를 삭제하고 X_train에 넣어주기
#drop은 삭제된 값을 반환하는게 아니라 삭제하고 남은 값을
#반환
# axis =0 행 (가로)
# axis =1 열(세로)
X_train = train.drop(['label'], axis =1)
X_test = test
y_train = train['label']

print(X_train.shape)
print(X_test.shape)
y_train
```

```python
from sklearn.neighbors import KNeighborsClassifier
# knn불러오기
knn = KNeighborsClassifier(n_neighbors = 5)
knn.fit(X_train, y_train)
y_test_pred = knn.predict(X_test)
answer['label']= y_test_pred
answer.to_csv('answer.csv', index=False)
answer
```
