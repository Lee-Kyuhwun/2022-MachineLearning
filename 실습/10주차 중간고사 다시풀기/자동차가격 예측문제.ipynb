```python
sample = pd.read_csv('/kaggle/input/2022-ml-midterm-p4/sample.csv')
train = pd.read_csv('/kaggle/input/2022-ml-midterm-p4/train.csv')
test = pd.read_csv('/kaggle/input/2022-ml-midterm-p4/test.csv')
```

```python
from sklearn import preprocessing
le = preprocessing.LabelEncoder()

train['model'] = le.fit_transform(train['model'])
test['model'] = le.fit_transform(test['model'])
```

```python
train_x = train.drop(['ID','price'],axis=1)
train_y = train['price']
test_x = test.drop(['ID'],axis=1)
```

```python
from sklearn.preprocessing import StandardScaler

sc = StandardScaler()
x_train_sc = sc.fit_transform(train_x)
x_test_sc = sc.transform(test_x)
```

```python
from sklearn.neighbors import KNeighborsRegressor

neigh = KNeighborsRegressor(n_neighbors=4,p=1)
neigh.fit(x_train_sc, train_y)
y_test_pred = neigh.predict(x_test_sc)
```

```python
sample['price']=y_test_pred
sample.to_csv('submit.csv', index=False)
```
