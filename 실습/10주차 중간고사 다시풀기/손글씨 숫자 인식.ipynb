```python
train = pd.read_csv('/kaggle/input/2022-ml-midterm-p2/train.csv')
test = pd.read_csv('/kaggle/input/2022-ml-midterm-p2/test.csv')
```

```python
train_x = train.drop(['label'], axis =1)
test_x = test
train_y = train['label']
```

```python
from sklearn.preprocessing import MinMaxScaler
scalar = MinMaxScaler()
scalar.fit(train_x)
x_train_std = scalar.transform(train_x)
x_test_std = scalar.transform(test_x)
```

```python
from sklearn.neighbors import KNeighborsClassifier
knn= KNeighborsClassifier()
knn.fit(x_train_std,train_y)
y_pred=  knn.predict(x_test_std)
```
