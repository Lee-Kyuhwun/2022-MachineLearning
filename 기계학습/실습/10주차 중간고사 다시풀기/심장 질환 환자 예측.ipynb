```python
train = pd.read_csv('/kaggle/input/2022-ml-midterm-p1/train.csv')
test = pd.read_csv('/kaggle/input/2022-ml-midterm-p1/test.csv')
```

```python
y_train = train['target']
x_train = train.drop(['target'],axis=1)
x_test = test
```

```python
y_train[y_train>0] = 1
```

```python
from sklearn.linear_model import LogisticRegression

log = LogisticRegression(solver='saga', max_iter=10000)
log.fit(x_train,y_train)
y_pred = log.predict(x_test)
```

```python
submit = pd.read_csv('/kaggle/input/2022-ml-midterm-p1/submit.csv')
submit['target']= y_pred
submit.to_csv('submit.csv',index =False)
```
