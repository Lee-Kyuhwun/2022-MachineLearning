```python
x_train = pd.read_csv('/kaggle/input/2022-ml-midterm-p3/train.csv')
x_test = pd.read_csv('/kaggle/input/2022-ml-midterm-p3/test.csv')
```

```python
y_train = x_train['price']
x_train = x_train.drop(['price','m'], axis =1)
x_test = x_test.drop(['m'],axis=1)
```

```python
from sklearn.preprocessing import StandardScaler

sc = StandardScaler()
x_train_sc = sc.fit_transform(x_train)
x_test_sc = sc.transform(x_test)
```

```python
from sklearn.neighbors import KNeighborsRegressor

neigh = KNeighborsRegressor(n_neighbors = 2).fit(x_train_sc, y_train)
y_test_pred = neigh.predict(x_test_sc)
```

```python
submit = pd.read_csv('/kaggle/input/2022-ml-midterm-p3/submit.csv')
submit['price'] = y_test_pred
submit.to_csv('submit.csv', index=False)
```
